## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].str.replace("$", "")
school_summary_df.head(2)
```

```python
school_summary_df["Per Student Budget"] = pd.to_numeric(school_summary_df["Per Student Budget"])
```

```python
school_summary_df["Per Student Budget"].dtype
```

```python
# Create per student budget bins
per_student_budget_bin = [0, 585, 630, 645, 700]
budget_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# use pd.cut to add column with appropriate corresponding bins
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], per_student_budget_bin, labels=budget_labels, include_lowest=True)
school_summary_df
```

```python
# groupby per student budget
scores_by_budget_avmath = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
scores_by_budget_avmath
```

```python
#% Pass MAth df column
scores_by_budget_percent_math = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Students Passing Math"]
scores_by_budget_percent_math
```

```python
#% Pass reading df column
scores_by_budget_avread = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Students Passing Reading"]
scores_by_budget_avread
```

```python
#% Pass all df column
scores_by_budget_percent_all = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Students Overall Passing"]
scores_by_budget_percent_all
```
